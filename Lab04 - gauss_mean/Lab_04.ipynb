Prática de Suavização de Imagens

Aplique o filtro de Média a imagem abaixo para diferentes tamanhos de máscaras.

3x3; 7x7; 3x3 aplicado três vezes sequenciais.

```python
img = cv2.imread("sta2.png", cv2.IMREAD_GRAYSCALE)
```

```python
def convolution(image, kernel):
    # Obter as dimensões da imagem de entrada e do kernel
    altura_imagem, largura_imagem = image.shape
    altura_kernel, largura_kernel = kernel.shape

    # Calcular o preenchimento necessário para a imagem de saída
    pad_altura = altura_kernel // 2
    pad_largura = largura_kernel // 2

    # Inicializar uma imagem de saída vazia
    output_image = np.zeros_like(image, dtype=np.float32)

    # Realizar a convolução
    for y in range(altura_imagem):
        for x in range(largura_imagem):
            # Extrair a região de interesse 
            out = image[max(0, y - pad_altura):min(altura_imagem, y + pad_altura + 1),
                        max(0, x - pad_largura):min(largura_imagem, x + pad_largura + 1)]

            if out.shape == kernel.shape:
                # Multiplicação e soma elemento a elemento
                output_image[y, x] = np.sum(out * kernel)

    return output_image.astype(np.uint8)
```

```python
def mean_filter(image, kernel_size, conv=None, flip=True):
    # Cria uma matrix de 1/kernel_size**2
    kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / (kernel_size**2)
    if flip:
        kernel = np.flipud(np.fliplr(kernel)) # 180
    #print(kernel)    
    if conv == None:
        # Convolução usando a função filter2D
        output_image = cv2.filter2D(image, -1, kernel)
    else:
        output_image = conv(image, kernel)

    return output_image

# kernel 3
output_image = mean_filter(img, 3)
# kernel 7
output_image1 = mean_filter(img, 7)
# kernel 3
output_image2 = mean_filter(output_image, 7)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(131), plt.imshow(img, cmap='gray', vmin=0, vmax=255), plt.title('Img')
plt.subplot(132), plt.imshow(output_image, cmap='gray', vmin=0, vmax=255), plt.title('Filtro de Média 3x3')
plt.subplot(133), plt.imshow(output_image1, cmap='gray', vmin=0, vmax=255), plt.title('Filtro de Média 7x7')
plt.show()


plt.imshow(output_image2, cmap='gray', vmin=0, vmax=255)
plt.title('3 Vezes sequenciais')
plt.show()
```

Aplique o filtro Gaussiano à imagem abaixo para diferentes valores de desvio padrão (σ) e tamanho de máscara.

σ = 1.0 (kernel size 5×5). σ = 2.0 (kernel size 9×9). σ = 4.0 (kernel size 15×15).

```python
img2 = cv2.imread("ben2.png", cv2.IMREAD_GRAYSCALE)
```

```python
def gaussian_filter(image, kernel_size, sigma, conv=None):
    # Cria o kernel gaussiano
    kernel = gaussian_kernel(kernel_size, sigma)
    
    if conv == None:
        output_image = cv2.filter2D(image, -1, kernel)
    else:
        output_image = conv(image, kernel)

    return output_image

def gaussian_kernel(kernel_size, sigma):
    kernel = np.zeros((kernel_size, kernel_size))

    center = kernel_size // 2

    #máscara Gaussiana
    for x in range(kernel_size):
        for y in range(kernel_size):
            kernel[x, y] = (1 / (2 * np.pi * sigma**2)) * np.exp(-((x - center)**2 + (y - center)**2) / (2 * sigma**2))

    #normalize a máscara para que a soma seja 1
    kernel = kernel / np.sum(kernel)
    
    return kernel
```

```python
# Aplica o filtro gaussiano à imagem
out5 = gaussian_filter(img2, 5, 1.0)
out9 = gaussian_filter(img2, 9, 2.0)
out15 = gaussian_filter(img2, 15, 4.0)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(131), plt.imshow(img2, cmap='gray', vmin=0, vmax=255), plt.title('Imagem Original')
plt.subplot(132), plt.imshow(out5, cmap='gray', vmin=0, vmax=255), plt.title('σ = 1.0')
plt.subplot(133), plt.imshow(out9, cmap='gray', vmin=0, vmax=255), plt.title('σ = 2.0')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(out15, cmap='gray', vmin=0, vmax=255), plt.title('σ = 4.0')
plt.show()
```

Aplique o filtro de Mediana para a imagem abaixo, que apresenta um ruído salt and pepper
3×3 filter; 7×7 filter

```python
img3 = cv2.imread("sta2noi2.png", cv2.IMREAD_GRAYSCALE)
```

```python
img_median_3x3 = cv2.medianBlur(img3, 3)

img_median_7x7 = cv2.medianBlur(img3, 7)

plt.figure(figsize=(12, 6))
plt.subplot(131), plt.imshow(img3, cmap='gray', vmin=0, vmax=255), plt.title('Imagem Original')
plt.subplot(132), plt.imshow(img_median_3x3, cmap='gray', vmin=0, vmax=255), plt.title('Filtro Mediana 3x3')
plt.subplot(133), plt.imshow(img_median_7x7, cmap='gray', vmin=0, vmax=255), plt.title('Filtro Mediana 7x7')
plt.show()
```
