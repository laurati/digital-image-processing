```python
import cv2
import numpy as np
from collections import deque
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("art8.png", cv2.IMREAD_GRAYSCALE);
plt.imshow(image, cmap='gray');
```

```python
plt.hist(np.squeeze(image));
```

```python
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        if image[i][j] == 1:
            image[i][j] = 0
        else:
            image[i][j] = 1
```

```python
def connected_components(imagem_bin):
    labels = np.zeros_like(imagem_bin)
    current_label = 2

    def is_valid(x, y):
        return 0 <= x < imagem_bin.shape[0] and 0 <= y < imagem_bin.shape[1]

    def label_component(x, y):
        stack = [(x, y)] 
        while stack:
            x, y = stack.pop()
            if not is_valid(x, y) or imagem_bin[x, y] == 0 or labels[x, y] != 0:
                continue
            labels[x, y] = current_label
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    if dx != 0 or dy != 0:
                        stack.append((x + dx, y + dy))

    for i in range(imagem_bin.shape[0]):
        for j in range(imagem_bin.shape[1]):
            if imagem_bin[i, j] == 1 and labels[i, j] == 0:
                label_component(i, j)
                current_label += 1

    return labels
```

```python
componentes_conectados = connected_components(image)

num_objetos = np.max(componentes_conectados) - 1


plt.figure(figsize=(12, 5))
plt.subplot(131)
plt.imshow(image, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(132)
plt.imshow(componentes_conectados, cmap='gray')
plt.title('Componentes Conectados')
plt.axis('off')

plt.subplot(133)
plt.imshow(componentes_conectados, cmap='nipy_spectral')
plt.title('Imagem Colorida')
plt.axis('off')

plt.show()

print(f'Número de objetos na imagem componentes conectados: {num_objetos}')
```

2. Calcule o número de objetos da imagem clc3.png. Note que antes será necessário binarizar a imagem. Para isso, considere um limiar (*threshold*) de 67, isto é, valores acima de 67 na imagem original recebem zero, e os abaixo recebem 1 (o que resulta na imagem clc3thr1.png).

```python
imagem_original = cv2.imread('clc3.png', cv2.IMREAD_GRAYSCALE)
limiar = 163
imagem_binarizada = (imagem_original < limiar).astype(np.uint8)
componentes_conectados = connected_components(imagem_binarizada)

num_objetos = np.max(componentes_conectados) - 1
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(131)
plt.imshow(imagem_original, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')


plt.subplot(132)
plt.imshow(imagem_binarizada, cmap='gray')
plt.title('Imagem Binarizada')
plt.axis('off')


plt.subplot(133)
plt.imshow(componentes_conectados, cmap='gray')
plt.title('Componentes Conectados')
plt.axis('off')

plt.show()

print(f'Número de objetos na imagem componentes conectados: {num_objetos}')
```
